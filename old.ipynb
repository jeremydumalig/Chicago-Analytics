## Head-to-Head

```python
season = pd.read_csv("season_overview.csv").iloc[:, 1:]

head_to_head = season[(season.Date == 'November 6, 2021')]

head_to_head = head_to_head.iloc[0:2, :]
head_to_head.loc[len(head_to_head)] = list(head_to_head.columns)

head_to_head.to_csv("head_to_head.csv")
head_to_head
```

## Scoring/Offensive/Miscellaneous Comparison

```python
season_averages = pd.read_csv("season_averages.csv")

traditional_comparisons = pd.concat([season_averages, head_to_head])

traditional_comparisons.to_csv("traditional_comparisons.csv")
traditional_comparisons
```
