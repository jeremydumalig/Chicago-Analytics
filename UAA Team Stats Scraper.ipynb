```python
from webdriver_manager.chrome import ChromeDriverManager
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import os

pd.options.mode.chained_assignment = None
# pd.set_option('display.max_rows', None)
# pd.set_option('display.max_colwidth', None)
```

```python
team_stats_url = "https://athletics.uchicago.edu/sports/mbkb/2021-22/boxscores/20211106_d1ei.xml?view=teamstats"

driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get(team_stats_url)
soup = BeautifulSoup(driver.page_source, 'html.parser')
driver.quit()
```

```python
df = pd.DataFrame()

stats = [i.text for i in soup.find_all('div', {'class': 'stats-box half'})[1].find_all('th')]
values = [i.text.strip() for i in soup.find_all('div', {'class': 'stats-box half'})[1].find_all('td')]

df['Stats'] = stats[3:-1]
df['Team 1'] = [values[i-1] for i in range(1, len(values), 2)]
df['Team 2'] = [values[i] for i in range(1, len(values), 2)]
df.columns = ['Stats', stats[1], stats[2]]

df
```

```python
plays = pd.read_csv("pbp_test.csv").iloc[:, 1:]

pd.concat([df, plays])
```
